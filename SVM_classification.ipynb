```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC


#load datasets
df=pd.read_csv(r"decision_tree2.csv")
df.head()
```

```python
sns.scatterplot(x="Age",y="Salary",data=df,hue="Purchased")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["Purchased"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
sv=SVC(kernel='linear') #kernel value=linear,poly,rbf,precomputed,sigmoid
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)*100
```

```python
sv.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=sv)
plt.show()
```
