```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import scipy.cluster.hierarchy as sc
from sklearn.cluster import KMeans
df=pd.read_csv(r"hierarchical_clustering_sample.csv")  
df.head()
```

```python
sns.pairplot(data=df)
plt.show()
# Create a KMeans instance with 3 clusters: model
```

```python
sc.dendrogram(sc.linkage(df, method='ward',metric='euclidean'))
model = KMeans(n_clusters=3, random_state=42)
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')      
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=3, linkage='ward')
# Fit model to samples
df["predict"]=ac.fit_predict(df)
```

```python
sns.pairplot(df, hue='predict', palette='Set1')
plt.show()
```
