```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.cluster import KMeans  

# Sample K-Means clustering data
data = {
    "x": [1.0, 1.5, 5.0, 8.0, 1.0, 9.0, 8.0, 10.0, 9.0, 3.0, 2.0, 2.2, 7.0, 6.5, 8.5, 1.2, 1.8, 2.0],
    "y": [2.0, 1.8, 8.0, 8.0, 0.6, 11.0, 2.0, 2.0, 3.0, 3.5, 2.5, 2.8, 9.0, 7.5, 7.8, 1.4, 0.8, 1.0]
}

df = pd.DataFrame(data)
df.head()
```

```python
sns.pairplot( data=df)
plt.show()
```

```python
wcss=[]
for i in range(1,15):
    km=KMeans(n_clusters=i, init='k-means++')
    wcss.append(km.fit(df).inertia_)
```

```python
plt.plot(range(1,15),wcss,marker='o')
plt.xlabel('Number of clusters',)
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmn=KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
df["predict"] = kmn.fit_predict(df)
```
