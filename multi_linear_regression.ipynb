```python
#import library==>

import pandas as pd
#import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#load dataset==>
df=pd.read_csv(r"multi linear.csv")
df.head(2)
```

```python
#check to null value present in dataset if present 
#than remove it
df.isnull().sum()
```

```python
df.info() #check is there any catagorical data
```

```python
df.drop(columns="ExtraParticipation",inplace=True) #remove catagorical dataset column
```

```python
df.info() #now there are no catagorical data
```

```python
df.head(4)
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["PapersPracticed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
mlr.score(x_test,y_test)*100
```
