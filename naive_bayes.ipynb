```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from mlxtend.plotting import plot_decision_regions
from sklearn.metrics import confusion_matrix,recall_score,f1_score,precision_score
df= pd.read_csv(r"confu_matt.csv")
df.head()
```

```python
sns.kdeplot(data=df["score"])
plt.show()
```

```python
sns.scatterplot(x="cgpa",y="score",data=df,hue="placed")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["placed"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100, gnb.score(x_train,y_train)*100
```

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100, mnb.score(x_train,y_train)*100
```

```python
br=BernoulliNB()
br.fit(x_train,y_train)
```

```python
br.score(x_test,y_test)*100, br.score(x_train,y_train)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy(),clf=gnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy(),clf=mnb)
plt.show()
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy(),clf=br)
plt.show()
```
