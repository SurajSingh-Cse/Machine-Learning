```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
df= pd.read_csv("Heart Prediction.csv")
df.head(2)
```

```python
df.drop(columns=["Gender","BloodPressure", "Cholesterol","HeartRate","QuantumPatternFeature"],inplace=True)
```

```python
df.head(2)
```

```python
sns.scatterplot(x="Age",y="HeartDisease",data=df)
plt.show()
```

```python
x=df[["Age"]]
y=df["HeartDisease"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[40]])
```

```python
sns.scatterplot(x="Age",y="HeartDisease",data=df)
sns.lineplot(x="Age",y=lr.predict(x),data=df,color="red")
plt.show()
```
