```python
#import library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

#load datasets
df=pd.read_csv(r"knn_regression_dataset.csv")
df.head()
```

```python
x=df.drop(columns="Salary")
y=df["Salary"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
knn=KNeighborsRegressor(n_neighbors=3,p=2)
knn.fit(x_train,y_train)
```

```python
knn.score(x_test,y_test)*100
```

```python
knn.score(x_train,y_train)*100
```

```python
#to check which is best fitting values 
for i in range(1,30):
    knn1=KNeighborsRegressor(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train)*100,knn1.score(x_test,y_test)*100)
```
