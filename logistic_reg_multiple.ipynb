```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
df= pd.read_csv("Heart Prediction.csv")
df.head(2)
```

```python
df.drop(columns=["QuantumPatternFeature","Gender","Cholesterol","HeartRate"],inplace=True)
```

```python
df.head(2)
```

```python
sns.scatterplot(x="Age",y="BloodPressure",data=df,hue="HeartDisease")
plt.show()
```

```python
x=df.iloc[:,:-1]
y=df["BloodPressure"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LogisticRegression(max_iter=10000,solver='saga')
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[68,105]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```
